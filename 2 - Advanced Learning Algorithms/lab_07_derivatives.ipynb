```python
from sympy import symbols, diff
```

```python
J = (3)**2
J_epsilon = (3 + 0.001)**2
k = (J_epsilon - J)/0.001    # difference divided by epsilon
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k:0.6f} ")
```

```python
J = (3)**2
J_epsilon = (3 + 0.000000001)**2
k = (J_epsilon - J)/0.000000001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
```

```python
J, w = symbols('J, w')
```

```python
J=w**2
J
```

```python
dJ_dw = diff(J,w)
dJ_dw
```

```python
dJ_dw.subs([(w,2)])  
```

```python
w, J = symbols('w, J')

J = 2 * w
print(J)

dJ_dw = diff(J,w)

print(dJ_dw)

dJ_dw.subs([(w,-3)])

J = 2*3
J_epsilon = 2*(3 + 0.001)
k = (J_epsilon - J)/0.001
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_dw ~= k = {k} ")
```

```python
J, w, x, y, b, m = symbols('J w x y b m')

J = ((w*x + b) - y)**2 / (2*m)

dJ_dw = diff(J, w)
dJ_db = diff(J, b)

dJ_dw, dJ_db

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# Carregar direto do GitHub
url = "https://raw.githubusercontent.com/lgreski/pokemonData/master/Pokemon.csv"
df = pd.read_csv(url)

print("Dataset")

display(df)

df.isnull().sum()

ocorrencias_type_1 = df['Type1'].value_counts()

plt.figure(figsize=(20, 10))
sns.barplot(x=ocorrencias_type_1.index, y=ocorrencias_type_1.values, hue=ocorrencias_type_1.index ,palette='viridis' , legend= False)
plt.title("Quantidade Tipo 1 por elemento")
plt.ylabel("Quantidade")
plt.show()

```

```python
media_ataque_por_geracao = df.groupby('Generation')['Attack'].mean() 


plt.figure(figsize=(8, 5))
sns.barplot(x=media_ataque_por_geracao.index, y=media_ataque_por_geracao.values, hue=media_ataque_por_geracao.index ,palette='viridis' , legend= False)
plt.title("Media de ataque por geraÃ§Ã£o")
plt.ylabel("Media")
plt.show()
```

```python
correlacao = df.corr(numeric_only= True)

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title("Matriz de CorrelaÃ§Ã£o")
plt.show()
```

```python
# 1. Calcular mÃ©dia de Total por tipo
forca_por_tipo = df.groupby('Type1')['Total'].mean().sort_values(ascending=False)

# 2. Visualizar
plt.figure(figsize=(12, 6))
sns.barplot(x=forca_por_tipo.index, y=forca_por_tipo.values, palette='viridis' , legend= False)
plt.title("ForÃ§a mÃ©dia por tipo")
plt.xticks(rotation=45)  # Para girar os nomes dos tipos
plt.show()
```
